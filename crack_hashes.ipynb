```
url = 'https://insidesherpa.s3.amazonaws.com/vinternships/companyassets/MBA4MnZTNFEoJZGnk/passwd_dump.txt'
response = requests.get(url) 
dump = response.text
dump = dump.split("\n") # formats the hashes
for line in dump: # files are iterable
    line = line[line.find(':')+1:]
    with open('/content/drive/My Drive/pass_crack/hash.txt', 'a') as f:
      f.write(line + "\n")
      f.close
```
